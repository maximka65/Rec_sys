3. Какие ограничения есть у ALS? (Тип информации, линейность/нелинейность факторов и т д)


Как я понял ALS находит линейную зависимость между оценкой(Explicit фактором), а это значит что, будь между оценками некая другая зависимость, ALS будет сильно ошибаться. Ну например 5 не сильно лучше чем 4, обе оценки люди ставят если удовлетворены услугой, но между 3 и 4 разница намного больше, 3 может означать очень высокий уровень неудовлетворенности(возникшую проблему) а 4 полную удовлетвореность(но не идеал). тем самым между 3 и 4 разница сильнее чем между 4 и 5. (как например в случае квадратичной зависимости)
